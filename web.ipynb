```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup
import asyncio
```

```python
urlList = []
nameList = []

def fetchurl(num):
    data = requests.get(f"https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/name/{num}/synonyms/XML")
    multiData = requests.get(f"https://pubchem.ncbi.nlm.nih.gov/rest/pug/substance/name/{num}/synonyms/XML")
    if str(data)=="<Response [200]>":
        cid = BeautifulSoup(data.content).find('cid')
        cid = str(cid).replace("</cid>", "")
        cid = cid.replace("<cid>", "")            
        url = "https://pubchem.ncbi.nlm.nih.gov/compound/"+cid
        
        syn =  BeautifulSoup(data.content).find_all('synonym')[0]
        syn = str(syn).replace("</synonym>", "")
        syn = syn.replace("<synonym>","")
        
        return([str(url),syn])
    
    elif str(multiData)=="<Response [200]>":
        sid = BeautifulSoup(multiData.content).find('sid')
        sid = str(sid).replace("</sid>", "")
        sid = sid.replace("<sid>", "")            
        url = "https://pubchem.ncbi.nlm.nih.gov/substance/"+sid
        
        syn =  BeautifulSoup(multiData.content).find('synonym')
        syn = str(syn).replace("</synonym>", "")
        syn = syn.replace("<synonym>","")
        
        return([str(url),syn])
    
    else: return(["",""])
```

```python
async def appendData(df,num):
    for i in range(len(df)):
        if str(num[i])!="nan":
            num[i] = str(num[i]).split(" (")[0]
            url,name = fetchurl(num[i])
            print(url,name)
            urlList.append(url)
            nameList.append(name)
            
        else: 
            urlList.append("")
            nameList.append("")
```
